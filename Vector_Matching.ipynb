Obtenemos el porcentaje de similitud mediante Jaccard


```python
def P_similarity(M_input, V_list, * , S_algorithm = "neiandli"):
  L_percentage = {}

  neiandli_percentage = []
  faith_percentage = []
  gower_legendre_percentage = []
  jaccard_percentage = []
  threew_jaccard_percentage = []
  sokal_sneath_I_percentage = []
  sokal_sneath_II_percentage = []
  sokal_michener_percentage = []
  roger_tanimoto_percentage = []
  dice_percentage = []
  intersection_percentage = []
  innerproduct_percentage =  []
  russell_rao_percentage = []
  ochiaI_I_percentage = []
  forbesi_percentage = []
  fossum_percentage = []
  sorgenfrei_percentage = []
  mountford_percentage = []
  otsuka_percentage = []
  mcconnaughey_percentage = []
  tarwid_percentage = []
  kulczynski_II_percentage = []
  driver_kroeber_percentage = []
  johnson_percentage = []
  dennis_percentage = []
  simpson_percentage = []
  braun_banquet_percentage = []
  fager_mcgowan_percentage = []
  forbes_II_percentage = []
  sokal_sneath_IV_percentage = []
  gower_percentage = []
  pearson_I_percentage = []
  pearson_II_percentage = []
  pearson_III_percentage = []
  pearson_heron_I_percentage = []
  pearson_heron_II_percentage = []
  sokal_sneath_III_percentage = []
  sokal_sneath_V_percentage = []
  cole_percentage = []
  stiles_percentage = []
  ochiaI_II_percentage = []
  yuleq_percentage = []
  yulew_percentage = []
  kulczynski_I_percentage = []
  tanimoto_percentage = []
  disperson_percentage = []
  hamann_percentage = []
  michael_percentage = []
  goodman_kruskal_percentage = []
  anderberg_percentage = []
  baroni_urbani_buser_I_percentage = []
  baroni_urbani_buser_II_percentage = []
  peirce_percentage = []
  eyraud_percentage = []
  tarantula_percentage = []
  ample_percentage = []

  def neiandli():
    neiandli_percentage.append(2 * A / 
      ((A + B) + (A + C)))
    L_percentage['neiandli'] = neiandli_percentage
  
  def jaccard():
    jaccard_percentage.append(A / 
      (A + B + C))
    L_percentage['jaccard'] = jaccard_percentage

  def threew_jaccard():
    threew_jaccard_percentage.append(3 * A / 
      (3 * A + B + C))
    L_percentage['threew_jaccard'] = threew_jaccard_percentage    

  def sokal_sneath_I():
    sokal_sneath_I_percentage.append(
      A / (A + 2 * B + 2 * C)
    )
    L_percentage['sokal_sneath_I'] = sokal_sneath_I_percentage

  def sokal_sneath_II():
    sokal_sneath_II_percentage.append(
      2  * (A + D) / 
        (2 * A +  B +  C + 2 * D)
    )
    L_percentage['sokal_sneath_II'] = sokal_sneath_II_percentage    

  def sokal_michener():
    sokal_michener_percentage.append(
      (A + D) / 
        (A +  B +  C + D)
    )
    L_percentage['sokal_michener'] = sokal_michener_percentage

  def roger_tanimoto():
    roger_tanimoto_percentage.append(
      (A + D) / 
        (A +  2 * (B +  C) + D)
    )
    L_percentage['roger_tanimoto'] = roger_tanimoto_percentage

  def gower_legendre():
    gower_legendre_percentage.append(
      (A + D) / 
        (A +  0.5 * (B +  C) + D)
    )
    L_percentage['gower_legendre'] = gower_legendre_percentage  

  def faith():
    faith_percentage.append(
      (A + 0.5 * D) / 
        (A +  B +  C + D)
    )
    L_percentage['faith'] = faith_percentage      

  def dice():
    dice_percentage.append(2 * A / 
      (2 * A + B + C))
    L_percentage['dice'] = dice_percentage

  def intersection():
    intersection_percentage.append(A)
    L_percentage['intersection'] = intersection_percentage

  def innerproduct():
    innerproduct_percentage.append(A + D)
    L_percentage['innerproduct'] = innerproduct_percentage
  
  def russell_rao():
    russell_rao_percentage.append(
      A / 
        (A +  B +  C + D)
    )
    L_percentage['russell_rao'] = russell_rao_percentage

  def ochiaI_I():
    ochiaI_I_percentage.append(
      A / pow((A + B) * (A + C), 0.5)
    )
    L_percentage['ochiaI_I'] = ochiaI_I_percentage

  def forbesi():
    forbesi_percentage.append(
      N * A / ((A + B) * (A + C))
    )
    L_percentage['forbesi'] = forbesi_percentage

  def fossum():
    fossum_percentage.append(
      (N * pow(A - 0.5, 2)) / ((A + B) * (A + C))
    )
    L_percentage['fossum'] = fossum_percentage

  def sorgenfrei():
    sorgenfrei_percentage.append(
      pow(A, 2) / ((A + B) * (A + C))
    )
    L_percentage['sorgenfrei'] = sorgenfrei_percentage

  def mountford():
    mountford_percentage.append(
      A / 0.5 * ((A * B) + (A * C)) 
        + (B * C)
    )  
    L_percentage['mountford'] = mountford_percentage

  def otsuka():
    otsuka_percentage.append(
      A / pow(((A + B) * (A + C)), 0.5)
    )
    L_percentage['otsuka'] = otsuka_percentage

  def mcconnaughey():
    mcconnaughey_percentage.append(
      (pow(A, 2) - B * C) 
        / ((A + B) * (A + C))
    )
    L_percentage['mcconnaughey'] = mcconnaughey_percentage

  def tarwid():
    tarwid_percentage.append(
      (N * A - ((A + B) * (A + C))) /
        (N * A + ((A + B) * (A + C)))
    )
    L_percentage['tarwid'] = tarwid_percentage

  def kulczynski_II():
    kulczynski_II_percentage.append(
      (A / 2) * (2 * A + B + C) /
        ((A + B) * (A + C))    
    )
    L_percentage['kulczynski_II'] = kulczynski_II_percentage

  def driver_kroeber():
    driver_kroeber_percentage.append(
      (A / 2) * ((1 / (A + B)) + (1 / (A + C)))    
    )
    L_percentage['driver_kroeber'] = driver_kroeber_percentage

  def johnson():
    johnson_percentage.append(
      (A / (A + B)) + (A / (A + C))    
    )
    L_percentage['johnson'] = johnson_percentage

  def dennis():
    dennis_percentage.append(
      ((A * D) - (B * C))  
        / np.sqrt(N * (A + B) * (A + C))  
    )
    L_percentage['dennis'] = dennis_percentage

  def simpson():
    simpson_percentage.append(
      (A / min(A + B, A + C))        
    )
    L_percentage['simpson'] = simpson_percentage

  def braun_banquet():
    braun_banquet_percentage.append(
      (A / max(A + B, A + C))        
    )
    L_percentage['braun_banquet'] = braun_banquet_percentage

  def fager_mcgowan():
    fager_mcgowan_percentage.append(
      (A / np.sqrt((A + B) * (A + C)))
        - (max(A + B, A + C) / 2)                    
    )
    L_percentage['fager_mcgowan'] = fager_mcgowan_percentage

  def forbes_II():
    forbes_II_percentage.append(
      (N * A - ((A + B) * (A + C))) / (N * min(A + B, A + C) - ((A + B) * (A + C)))
    )
    L_percentage['forbes_II'] = forbes_II_percentage

  def sokal_sneath_IV():
    sokal_sneath_IV_percentage.append(
      ((A / (A + B)) + (A / (A + C))
        + (D / (B + D)) + (D / (B + D))) / 4     
    )
    L_percentage['sokal_sneath_IV'] = sokal_sneath_IV_percentage

  def gower():
    gower_percentage.append(
      (A + D) / np.sqrt((A + B) * (A + C) * (B + D) * (C + D))
    )
    L_percentage['gower'] = gower_percentage

  def pearson_I():
    pearson_I_percentage.append(
      X_2
    )
    L_percentage['pearson_I'] = pearson_I_percentage

  def pearson_II():
    pearson_II_percentage.append(
      np.sqrt(pow(X_2, 2) / (N + pow(X_2, 2))) 
    )
    L_percentage['pearson_II'] = pearson_II_percentage

  def pearson_III():
    pearson_III_percentage.append(
      pow(P / (N + P), 0.5)    
    )
    L_percentage['pearson_III'] = pearson_III_percentage

  def pearson_heron_I():
    pearson_heron_I_percentage.append(
      P
    )              
    L_percentage['pearson_heron_I'] = pearson_heron_I_percentage

  def pearson_heron_II():
    pearson_heron_II_percentage.append(
      np.cos(np.pi * np.sqrt(B * C) / (np.sqrt(A * D) + np.sqrt(B * C)))
    )
    L_percentage['pearson_heron_II'] = pearson_heron_II_percentage

  def sokal_sneath_III():
    try:
      sokal_sneath_III_percentage.append(
        (A + D) / (B + C)
      )
      L_percentage['sokal_sneath_III'] = sokal_sneath_III_percentage  

    except ZeroDivisionError:
      raise

  def sokal_sneath_V():
    sokal_sneath_V_percentage.append(
      A * D / ((A + B) * (A + C) * (B + D) * pow(C + D, 0.5))  
    )
    L_percentage['sokal_sneath_V'] = sokal_sneath_V_percentage

  def cole():
    cole_percentage.append(
      np.sqrt(2) * ((A * D) - (B * C)) / np.sqrt(pow(A * D - B * C, 2) -
        (A + B) * (A + C) * (B + D) * (C + D)) 
    )
    L_percentage['cole'] = cole_percentage

  def stiles():
    stiles_percentage.append(
      np.log10(N * pow(abs((A * D) - (B * C)) - (N / 2), 2) / ((A + B) * (A + C) *
                                                             (B + D) * (C + D)))  
    )
    L_percentage['stiles'] = stiles_percentage

  def ochiaI_II():
    ochiaI_II_percentage.append(
      A * D / pow(((A + B) * (A + C)
        * (B + D) * (C + D)), 0.5)
    )
    L_percentage['ochiaI_II'] = ochiaI_II_percentage

  def yuleq():
    yuleq_percentage.append(
      ((A * D) - (B * C)) / ((A * D) + (B * C))   
    )
    L_percentage['yuleq'] = yuleq_percentage

  def yulew():
    yulew_percentage.append(
      (np.sqrt(A * D) - np.sqrt(B * C)) / (np.sqrt(A * D) + np.sqrt(B * C))     
    )
    L_percentage['yulew'] = yulew_percentage

  def kulczynski_I():
    try:
      kulczynski_I_percentage.append(
        A / (B + C)    
      )
      L_percentage['kulczynski_I'] = kulczynski_I_percentage
    except ZeroDivisionError:
      raise

  def tanimoto():
    tanimoto_percentage.append(
      A / ((A + B) + (A + C) - A) 
    )
    L_percentage['tanimoto'] = tanimoto_percentage

  def disperson():
    disperson_percentage.append(
      ((A * D) - (B * C)) / pow(A + B + C + D, 2)
    )
    L_percentage['disperson'] = disperson_percentage

  def hamann():
    hamann_percentage.append(
      ((A + D) - (B + C)) / (A + B + C + D) 
    )
    L_percentage['hamann'] = hamann_percentage

  def michael():
    michael_percentage.append(
      4 * ((A * D) - (B * C)) /
        (pow(A + D, 2) + pow(B + C, 2))
    )
    L_percentage['michael'] = michael_percentage

  def goodman_kruskal():
    goodman_kruskal_percentage.append(
      (G - G_2) / (2 * N - G_2)
    )
    L_percentage['goodman_kruskal'] = goodman_kruskal_percentage

  def anderberg():
    anderberg_percentage.append(
      (G - G_2) / (2 * N) 
    )
    L_percentage['anderberg'] = anderberg_percentage

  def baroni_urbani_buser_I():
    baroni_urbani_buser_I_percentage.append(
      (np.sqrt(A * D) + A) / (np.sqrt(A * D) + A + B + C)
    )
    L_percentage['baroni_urbani_buser_I'] = baroni_urbani_buser_I_percentage

  def baroni_urbani_buser_II():
    baroni_urbani_buser_II_percentage.append(
      (np.sqrt(A * D) + A - (B + C)) / (np.sqrt(A * D) + A + B + C)
    )
    L_percentage['baroni_urbani_buser_II'] = baroni_urbani_buser_II_percentage

  def peirce():
    try:
      peirce_percentage.append(
        (A * B + B * C) / (A * B + (2 * (B * C)) + C * D)
      )
      L_percentage['peirce'] = peirce_percentage
    except ZeroDivisionError:
      raise

  def eyraud():
    eyraud_percentage.append(
      pow(N, 2) * ((N * A) - (A + B) * (A + C)) / ((A + B) * (A + C) * (B + D) * (C + D)) 
    )
    L_percentage['eyraud'] = eyraud_percentage

  def tarantula():
    try:
      tarantula_percentage.append(
        A * (C + D) / (C * (A + B))
      )
      L_percentage['tarantula'] = tarantula_percentage
    except ZeroDivisionError:
      raise

  def ample():
    try:
      ample_percentage.append(
        abs(A * (C + D) / (C * (A + B)))
      )             
      L_percentage['ample'] = ample_percentage
    except ZeroDivisionError:
      raise

  switch_algorithm = {
      'neiandli': neiandli,
      'faith': faith,
      'gower_legendre': gower_legendre,
      'jaccard': jaccard,
      'roger_tanimoto': roger_tanimoto,
      'sokal_michener': sokal_michener,
      'sokal_sneath_I': sokal_sneath_I,
      'sokal_sneath_II': sokal_sneath_II,
      'threew_jaccard': threew_jaccard, 
      'dice': dice,
      'intersection': intersection,
      'innerproduct': innerproduct,
      'russell_rao': russell_rao,
      'ochiaI_I': ochiaI_I,
      'forbesi': forbesi,
      'fossum': fossum,
      'sorgenfrei': sorgenfrei,
      'mountford': mountford,
      'otsuka': otsuka,
      'mcconnaughey': mcconnaughey,
      'tarwid': tarwid,
      'kulczynski_II': kulczynski_II,
      'driver_kroeber': driver_kroeber,
      'johnson': johnson,
      'dennis': dennis,
      'simpson': simpson,
      'braun_banquet': braun_banquet,
      'fager_mcgowan': fager_mcgowan,
      'forbes_II': forbes_II,
      'sokal_sneath_IV': sokal_sneath_IV,
      'gower': gower,
      'pearson_I': pearson_I,
      'pearson_II': pearson_II,
      'pearson_III': pearson_III,
      'pearson_heron_I': pearson_heron_I,
      'pearson_heron_II': pearson_heron_II,
      'sokal_sneath_III': sokal_sneath_III,
      'sokal_sneath_V': sokal_sneath_V,
      'cole': cole,
      'stiles': stiles,
      'ochiaI_II': ochiaI_II,
      'yuleq': yuleq,
      'yulew': yulew,
      'kulczynski_I': kulczynski_I,
      'tanimoto': tanimoto,
      'disperson': disperson,
      'hamann': hamann,
      'michael': michael,
      'goodman_kruskal': goodman_kruskal,
      'anderberg': anderberg,
      'baroni_urbani_buser_I': baroni_urbani_buser_I,
      'baroni_urbani_buser_II': baroni_urbani_buser_II,
      'peirce': peirce,
      'eyraud': eyraud,
      'tarantula': tarantula,
      'ample': ample
    }

  for item in V_list:
    # Generamos indices boleanos para acceder a los elementos que cumplan la condicional
    L_A = np.logical_and(M_input == 1, item == 1 ) #(i == 1) & (j == 1)
    L_B = np.logical_and(M_input == 0, item == 1) #(i == 0) & (j == 1)
    L_C = np.logical_and(M_input == 1, item == 0) #(i == 1) & (j == 0)
    L_D = np.logical_and(M_input == 0, item == 0 ) #(i == 0) & (j == 0)

    A = M_input[L_A].size
    B = M_input[L_B].size
    C = M_input[L_C].size
    D = M_input[L_D].size

    N = A +  B +  C + D
    X_2 = N * pow(A * D - B * C, 2) / ((A + B) * (A + C) * (C + D) * (B + D))
    P = ((A * D) - (B * C)) / np.sqrt((A + B) * (A + C) * (B + D) * (C + D))
    G = max(A, B) + max(C, D) + max(A, C) + max(B, D)

    G_2 = max(A + C, B + D) + max(A + B, C + D)     
    
    switch_algorithm.get('neiandli')() 
    #switch_algorithm.get('faith')()
    #switch_algorithm.get('gower_legendre')() P.C.
    switch_algorithm.get('jaccard')()
    #switch_algorithm.get('roger_tanimoto')() P.C
    #switch_algorithm.get('sokal_sneath_I')() P.C
    #switch_algorithm.get('sokal_sneath_II')() P.C 
    switch_algorithm.get('sokal_michener')()
    #switch_algorithm.get('threew_jaccard')() P.C.
    switch_algorithm.get('dice')()
    #switch_algorithm.get('intersection')()
    #switch_algorithm.get('innerproduct')()
    #switch_algorithm.get('russell_rao')() I.
    switch_algorithm.get('ochiaI_I')()
    #switch_algorithm.get('forbesi')()
    #switch_algorithm.get('fossum')()
    #switch_algorithm.get('sorgenfrei')() P.C.
    #switch_algorithm.get('mountford')()
    switch_algorithm.get('otsuka')()
    #switch_algorithm.get('mcconnaughey')() I.
    #switch_algorithm.get('tarwid')()
    switch_algorithm.get('kulczynski_II')()
    switch_algorithm.get('driver_kroeber')()
    #switch_algorithm.get('johnson')()
    #switch_algorithm.get('dennis')()
    #switch_algorithm.get('simpson')()
    switch_algorithm.get('braun_banquet')()
    #switch_algorithm.get('fager_mcgowan')()
    #switch_algorithm.get('forbes_II')()
    switch_algorithm.get('sokal_sneath_IV')()
    #switch_algorithm.get('gower')()
    #switch_algorithm.get('pearson_I')()
    #switch_algorithm.get('pearson_II')()
    #switch_algorithm.get('pearson_III')()
    #switch_algorithm.get('pearson_heron_I')()
    #switch_algorithm.get('pearson_heron_II')()
    #switch_algorithm.get('sokal_sneath_III')() I.
    #switch_algorithm.get('sokal_sneath_V')()
    #switch_algorithm.get('cole')()
    #switch_algorithm.get('stiles')()
    switch_algorithm.get('ochiaI_II')()
    #switch_algorithm.get('yuleq')()
    #switch_algorithm.get('yulew')()
    #switch_algorithm.get('kulczynski_I')()
    switch_algorithm.get('tanimoto')()
    #switch_algorithm.get('disperson')()
    #switch_algorithm.get('hamann')() I. 
    #switch_algorithm.get('michael')()
    #switch_algorithm.get('goodman_kruskal')() I.
    #switch_algorithm.get('anderberg')()
    switch_algorithm.get('baroni_urbani_buser_I')()
    #switch_algorithm.get('baroni_urbani_buser_II')() I.
    #switch_algorithm.get('peirce')()
    #switch_algorithm.get('eyraud')()
    #switch_algorithm.get('tarantula')()
    #switch_algorithm.get('ample')() 

  return L_percentage 
  #return {S_algorithm: L_percentage.get(S_algorithm)} # Regresa un algoritmo especifico 
```
